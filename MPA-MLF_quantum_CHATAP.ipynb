```python
# Implement a half adder in quantum circuit:
```

```python
# Create quantum circuit with 2 qubits and 2 classical bits
qc = QuantumCircuit(4, 2)
qc.x(0)
qc.x(1)
qc.cx(1,2)
qc.cx(0,2)  # CNOT controlled by qubit 0 and targeting qubit 1
qc.ccx(0,1,3)
qc.measure(2,0)
qc.measure(3,1)
display(qc.draw())     # display a drawing of the circuit
job = sim.run(qc)      # run the experiment
result = job.result()  # get the results
# interpret the results as a "counts" dictionary
result.get_counts()
```

```python
from qiskit.visualization import plot_histogram
plot_histogram(result.get_counts())
```

```python
# Check that Hadamard gate changes the state to + state with uncertainty:
```

```python
qc2 = QuantumCircuit(2, 2)
qc2.h(0)
qc2.measure(0,0)
qc2.measure(1,1)
display(qc2.draw())     # display a drawing of the circuit
job2 = sim.run(qc2)      # run the experiment
result2 = job2.result()  # get the results
# interpret the results as a "counts" dictionary
result2.get_counts()
```

```python
backend = Aer.get_backend('statevector_simulator')
qc2.x(1)
job3= backend.run(qc2)
result3=job3.result()
outputstate = result3.get_statevector(qc2,decimals=3)
print(outputstate)
```

```python
#state 1
qc_1=QuantumCircuit(1)
state_1 = [1/sqrt(2), 1/sqrt(2)]# Define first state as (|0> + |1>)/sqrt(2) instead of |1>
qc_1.initialize(state_1, 0)# Apply initialisation operation to the 0th qubit
qc_1.draw()
plot_bloch_multivector(state_1)
```

```python
#state 2
qc_2=QuantumCircuit(1)
state_2 = [1/sqrt(2), -1j/sqrt(2)]# Define second state as (|0> - i|1>)/sqrt(2) instead of |1>
qc_2.initialize(state_2, 0)# Apply initialisation operation to the 0th qubit
qc_2.draw()
plot_bloch_multivector(state_2)
```

```python
#state 3 
qc_3=QuantumCircuit(1)
state_4 = [2/sqrt(5), 1/sqrt(5)]# Define third state as (|0> + 2|1>)/sqrt(5) instead of |1>
qc_3.initialize(state_4, 0)# Apply initialisation operation to the 0th qubit
qc_3.draw()
plot_bloch_multivector(state_4)
```

```python
#State 4
qc_4=QuantumCircuit(1)
state_4 = [2/sqrt(5), -1j/sqrt(5)]# Define fourth state as [2/sqrt(5), -1j/sqrt(5)] instead of |1>
qc_4.initialize(state_4, 0)# Apply initialisation operation to the 0th qubit
qc_4.draw()
plot_bloch_multivector(state_4)
```
